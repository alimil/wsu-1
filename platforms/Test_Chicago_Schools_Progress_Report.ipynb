```python
import pandas as pd


chicago_schools = pd.read_csv("https://storage.googleapis.com/mbcc/datasets/chigago_schools_progress_reports.csv")

# import the regular expression library to help us scrub the column names
import re


# columns names have comma and spaces so we will get rid of them
new_col_names = []

# strip() removes extra spaces around the value and also replace comma(,) and inner space with _ using regex. For using regex we import re
for column in chicago_schools.columns:
    column = column.strip()
    new_col = re.sub(r"[(|)| |,]", "_", column)
    new_col_names.append(new_col.lower())
    
chicago_schools.columns = new_col_names


```
